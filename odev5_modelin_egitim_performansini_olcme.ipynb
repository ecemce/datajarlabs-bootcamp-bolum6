## 1. Hava durumu modeli

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from scipy.stats.mstats import winsorize
import warnings
```

```python
df = pd.read_csv('weather.csv')
df.head()
```

```python
Y = df['Temperature (C)'] - df['Apparent Temperature (C)']
X = df[['Humidity','Wind Speed (km/h)']]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

R-kare ve ayarlanmış R-kare değerleri birbirine eşit ve 0.288'dir.R-kare değerinin overfitting riski taşımadan 1' yakın olmasını isteriz
bu sebepten 0.288 değeri bizim için düşük bir sonuçtur

### Yeni modelimiz

```python
df["humidity_windspeed"]=df["Humidity"]*df['Wind Speed (km/h)']
X = df[["Humidity","Wind Speed (km/h)","humidity_windspeed"]]

X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

R-kare ve ayarlanmış R-kare sonuçlarımız yine aynı ve 0.341 olarak değişmekte olup bir artış söz konusudur.
2. modelimiz ilk modele göre daha iyi bir konumdadır.

#### Görünürlük eklenmiş model

```python
X = df[['Humidity','Wind Speed (km/h)', 'Visibility (km)']]

X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

R-kare değerimiz 0.304 ayarlanmış R-kare değeri 0.303'tür.R-kare değeri model 3'te model 2'ye göre 0.037'lik bir düşüşe sahiptir.
Yani R-kare açısında model 2 daha iyi gözükmektedir.

AIC ve BIC değerleri açısından;
1. model için AIC :3.409e+05 BIC:3.409e+05
2. model için AIC :3.334e+05 BIC:3.334e+05
3. model için AIC :3.388e+05 BIC:3.388e+05
AIC ve BIC değeri en düşük olan 2. model en iyi modeldir.Sonrasında 3.model ve son sırada 1. model gelmektedir.
