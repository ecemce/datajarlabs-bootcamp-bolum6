### 2. Hava durumu modeli

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from scipy.stats.mstats import winsorize
import warnings
```

```python
df = pd.read_csv('weather.csv')
df.head()
```

```python
Y = df['Temperature (C)'] - df['Apparent Temperature (C)']
X = df[['Humidity','Wind Speed (km/h)']]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

P değerleri incelendiğinde nem ve rüzgar hızı değişkenlerinin hedef değişken üzerinde istatiksel olarak anlamlı olduğunu söyleyebiliriz.

Ayrıca nem ve rüzgar hızı hedef değişken üzerinde olumlu bir etkiye sahiptir.Yani nem ve rüzgar hızı arttığında sıcaklık da artmaktadır.

```python
df["humidity_windspeed"]=df["Humidity"]*df['Wind Speed (km/h)']
X = df[["Humidity","Wind Speed (km/h)","humidity_windspeed"]]

X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

Modele etkileşimli özelliğimizi dahil ettiğimizde de özellikler yine istatistiksel olarak anlamlıdır.

Değişen modelimizde nem ve rüzgar hızının hedef değişken üzerindeki etkisi değişmiştir , nem ve rüzgar hızı arttığı zaman hissedilen sıcaklık azalmaktadır.

### 3. Ev fiyatları modeli

```python
df3 = pd.read_csv('train.csv')
df3.head()
```

```python
df3 = df3.drop(['Alley', 'PoolQC', 'Fence','MiscFeature','LotFrontage', 'FireplaceQu' ], axis = 1)
df3 = df3.dropna()
```

```python
Y = df3["SalePrice"]
X = df3[["MSSubClass","LotArea","MoSold","PoolArea"]]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

P değerlerini incelediğimizde Mosold ve MSSubClass değişkenlerinin hedef değişken üzerinde anlamsız olduğunu görmekteyiz.Bu değişkenleri çıkararak devam edelim.

```python
X = X.drop(["MoSold","MSSubClass"], axis = 1)
X.head()
```

```python
X = sm.add_constant(X)
resultss = sm.OLS(Y,X).fit()
resultss.summary()
```

Hedef değişken üzerinde etkisi en fazla olan özelliğimiz LotArea değişkenidir.

PoolArea değişkeninin hedef değişken üzerindeki etkisi LotArea'ya göre daha azdır.Fakat yinede istatistiksel olarak anlamlıdır.
